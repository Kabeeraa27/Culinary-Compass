# ENCODING

```python
# Label encoding binary categorical columns
le = LabelEncoder()
df['Has Table booking'] = le.fit_transform(df['Has Table booking'])
df['Has Online delivery'] = le.fit_transform(df['Has Online delivery'])
df['Is delivering now'] = le.fit_transform(df['Is delivering now'])
df['Switch to order menu'] = le.fit_transform(df['Switch to order menu'])
df['Rating color'] = le.fit_transform(df['Rating color'])
df['Rating text'] = le.fit_transform(df['Rating text'])

# One-hot encoding multi-class categorical columns
df = pd.get_dummies(df, columns=['Country Code', 'City', 'Currency'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#VECTORIZING CUISINES TYPES
tfidf = TfidfVectorizer(stop_words='english')
df['Cuisines'] = df['Cuisines'].fillna('')
tfidf_matrix = tfidf.fit_transform(df['Cuisines'])

#COSINE SIMILARITY
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

# RECOMMENDATION FUNCTION

```python
def get_recommendations(restaurant_index, cosine_sim=cosine_sim):
    
    sim_scores = list(enumerate(cosine_sim[restaurant_index]))          # Get Similarity Scores (SS)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)   # Sort restaurants based on SS
    sim_scores = sim_scores[1:16]                                       # Top 15 most Similar restaurants
    restaurant_indices = [i[0] for i in sim_scores]                     # Indices
    
    # Extract city names
    recommended_restaurants = df.iloc[restaurant_indices].copy()
    city_columns = [col for col in df.columns if col.startswith('City_')]
    for idx in recommended_restaurants.index:
        city = df.loc[idx, city_columns].idxmax().replace('City_', '')
        recommended_restaurants.loc[idx, 'City'] = city
    
    return recommended_restaurants[['Restaurant Name', 'Aggregate rating', 'City', 'Cuisines','Price range','Votes', 'Has Table booking', 'Has Online delivery', 'Address' ]]
```

- Sample user preferences

```python
user_preferences = {
    'Cuisines': 'Thai',
    'City': 'Dalton',
    'Price range': 2,
    'Aggregate rating': 4.2,
    'Has Table booking': 0,
    'Has Online delivery': 0,
    }
```

- Filter restaurants based on user preferences

```python
city_column = f"City_{user_preferences['City']}"
if city_column not in df.columns:
    raise ValueError(f"City '{user_preferences['City']}' not found in the dataset.")

filtered_df = df[
    (df['Cuisines'].str.contains(user_preferences['Cuisines'])) &
    (df[city_column] == 1) &
    (df['Price range'] <= user_preferences['Price range']) &
    (df['Aggregate rating'] >= user_preferences['Aggregate rating']) &
    (df['Has Table booking'] == user_preferences['Has Table booking']) &
    (df['Has Online delivery'] == user_preferences['Has Online delivery'])
]
```

- Recommend restaurants

```python
if filtered_df.empty:
    print("No restaurants found matching the user preferences.")
else:
    example_restaurant_index = filtered_df.index[0]
    recommendations = get_recommendations(example_restaurant_index)
    recommendations_df = pd.DataFrame(recommendations)
    recommendations_df
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import tkinter as tk
from tkinter import ttk
```

- FILTER FUNCTION

```python
def filter_restaurants():
    user_preferences = {
        'Cuisines': cuisines_entry.get(),
        'City': city_entry.get(),
        'Price range': int(price_range_entry.get()),
        'Aggregate rating': float(aggregate_rating_entry.get()),
        'Has Table booking': table_booking_var.get(),
        'Has Online delivery': online_delivery_var.get()
    }

    city_column = f"City_{user_preferences['City']}"
    if city_column not in df.columns:
        result_tree.delete(*result_tree.get_children())
        result_tree.insert('', 'end', values=(f"City '{user_preferences['City']}' not found in the dataset.",))
        return

    filtered_df = df[
        (df['Cuisines'].str.contains(user_preferences['Cuisines'], case=False)) &
        (df[city_column] == 1) &
        (df['Price range'] <= user_preferences['Price range']) &
        (df['Aggregate rating'] >= user_preferences['Aggregate rating']) &
        (df['Has Table booking'] == user_preferences['Has Table booking']) &
        (df['Has Online delivery'] == user_preferences['Has Online delivery'])
    ]

    result_tree.delete(*result_tree.get_children())
    
    if filtered_df.empty:
        result_tree.insert('', 'end', values=("No restaurants found matching the user preferences.",))
    else:
        example_restaurant_index = filtered_df.index[0]
        recommendations = get_recommendations(example_restaurant_index)
        for _, row in recommendations.iterrows():
            result_tree.insert('', 'end', values=list(row))
            
```
