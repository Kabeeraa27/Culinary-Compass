# LOCATIONWISE RESTAURANT STATS

```python
print(f"Total Number of Cities in this Dataset: {df['City'].nunique()} ")
```

```python
# GROUP BY CITY
city_group = df.groupby('City').size().reset_index(name='Restaurant Count')

# TOP 10 CITIES BY RESTAURANT COUNT
top_cities = city_group.sort_values(by='Restaurant Count', ascending=False).head(10)
print("TOP 10 CITIES BY RESTAURANT COUNT:")
top_cities
```

```python
# Average rating by city
avg_rating_by_city = df.groupby('City')['Aggregate rating'].mean().reset_index(name='Average Rating')

# Average price range by city
avg_price_by_city = df.groupby('City')['Price range'].mean().reset_index(name='Average Price Range')

# Number of different cuisines by city
cuisines_by_city = df.groupby('City')['Cuisines'].nunique().reset_index(name='Unique Cuisines')

city_stats = top_cities.merge(avg_rating_by_city, on='City')
city_stats = city_stats.merge(avg_price_by_city, on='City')
city_stats = city_stats.merge(cuisines_by_city, on='City')

print("CITY STATISTICS:")
city_stats
```

```python
# Most common cuisine by city, # Total votes by city
total_votes_by_city = df.groupby('City')['Votes'].sum().reset_index(name='Total Votes')
city_stats = city_stats.merge(total_votes_by_city, on='City')

city_stats
```

```python
def show_city_stats():
    city = city_var.get()
    city_df = df[df['City'] == city]

    # Calculating statistics
    restaurant_count = len(city_df)
    avg_rating = city_df['Aggregate rating'].mean()
    avg_rating_rounded = round(avg_rating, 4)  
    avg_price_range = city_df['Price range'].mean()
    avg_price_range_rounded = round(avg_price_range, 4)  
    unique_cuisines = city_df['Cuisines'].nunique()
    total_votes = city_df['Votes'].sum()

    # Most voted restaurant and its votes
    most_voted = city_df.loc[city_df['Votes'].idxmax(), 'Restaurant Name']
    votes_most_voted = city_df.loc[city_df['Votes'].idxmax(), 'Votes']

    # Most cuisines offered by a restaurant and the count
    most_cuisines_restaurant = city_df.loc[city_df['Cuisines'].apply(lambda x: len(x.split(','))).idxmax(), 'Restaurant Name']
    cuisines_count = city_df['Cuisines'].apply(lambda x: len(x.split(','))).max()

    # Most affordable and expensive restaurants
    most_affordable = city_df.loc[city_df['Price range'].idxmin(), 'Restaurant Name']
    most_expensive = city_df.loc[city_df['Price range'].idxmax(), 'Restaurant Name']

    # Get price range for most affordable and expensive restaurants
    price_range_most_affordable = city_df.loc[city_df['Price range'].idxmin(), 'Price range']
    price_range_most_expensive = city_df.loc[city_df['Price range'].idxmax(), 'Price range']

    # Update Treeview with new data
    result_tree.delete(*result_tree.get_children())
    result_tree.insert('', 'end', values=[
        city, restaurant_count, avg_rating_rounded, avg_price_range_rounded, unique_cuisines, total_votes,
        f"{most_voted} (V: {votes_most_voted})", 
        f"{most_cuisines_restaurant} (No.: {cuisines_count})", 
        f"{most_affordable} (PR: {price_range_most_affordable})", 
        f"{most_expensive} (PR: {price_range_most_expensive})"
    ])

root = tk.Tk()
root.title("Restaurant Statistics by City")

# City Dropdown
cities = sorted(df['City'].unique())
city_var = tk.StringVar(root)
city_var.set(cities[0])
city_dropdown = ttk.Combobox(root, textvariable=city_var, values=cities)
city_dropdown.grid(row=0, column=0, padx=10, pady=10)

# Button
ttk.Button(root, text="Show Stats", command=show_city_stats).grid(row=0, column=1, padx=10, pady=10)

# Treeview for displaying stats
columns = ['City', 'Restaurant Count', 'Average Rating', 'Average Price Range', 'Unique Cuisines', 
           'Total Votes', 'Most Voted Restaurant', 'Most Cuisines Restaurant', 
           'Most Affordable Restaurant', 'Most Expensive Restaurant']
result_tree = ttk.Treeview(root, columns=columns, show='headings', height=5)

column_widths = [150, 120, 100, 120, 120, 100, 200, 200, 200, 200]  
for col, width in zip(columns, column_widths):
    result_tree.heading(col, text=col)
    result_tree.column(col, anchor='center', width=width, minwidth=90)

result_tree.grid(row=1, column=0, columnspan=2, padx=10, pady=10)

root.mainloop()
```
